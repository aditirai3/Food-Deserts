```python
import pandas as pd
```

# Data loading and merging
* Read the food deserts csv from USDA and the csv with corresponding Lat Long info, downloaded from Census
* Merge both csv's

```python
main = pd.read_csv("DataDownload.csv")
```

```python
main
```

```python
latlong = pd.read_csv("2020_Gaz_tracts_national.csv")
```

```python
latlong
```

* **Rename GEOID column, which is the census tract id, in order to create a commom column for merging**

* **Drop irrelevant columns**

```python
latlong.rename(columns = {'GEOID':'CensusTract'}, inplace = True)
```

```python
latlong = latlong.drop(['USPS', 'ALAND', 'AWATER', 'ALAND_SQMI', 'AWATER_SQMI'], axis=1)
```

```python
latlong.head()
```

```python
final_csv = pd.merge(main, latlong, how='left', on='CensusTract')
```

# Data cleaning

* Rename Lat Long columns
* Check for and drop all rows that contain NaN values
* Export the final dataframe to a csv file

```python
final_csv.rename(columns = {'INTPTLAT': 'Latitude', 'INTPTLONG': 'Longitude'}, inplace=True)
```

```python
final_csv.head()
```

```python
print(final_csv.Latitude.isnull())
```

```python
print(final_csv.Longitude .isnull())
```

```python
final_csv = final_csv.dropna()
```

```python
final_csv.to_csv('combined.csv', header=True)
```
