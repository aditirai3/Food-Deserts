```python
# Dependencies
import requests
import json
import gmaps
import pandas as pd
```

```python
Wikiurl = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
```

```python
# use read_html to scrape out tables and select the relevant table
tables = pd.read_html(Wikiurl)[0]
tables.head()
```

```python
tables.columns = tables.columns.droplevel()
```

```python
tables.rename(columns = {'Flag, name andpostal abbreviation[12]':'State'}, inplace = True)
```

```python
tables.head()
```

```python
tables = tables.drop(columns = ['Flag, name andpostal abbreviation[12].1', 'Capital', 'Largest[16]', 'Ratification oradmission[C]', 'Population[D][14]', 'mi2', 'km2', 'Numberof Reps.'])
```

```python
tables.at[16, 'State']= "Kentucky"
tables.at[20, 'State']= "Massachusetts"
tables.at[37, 'State']= "Pennsylvania"
tables.at[45, 'State']= "Virginia"
```

```python
tables
```

```python
tables["Latitude"] = ""
tables["Longitude"] = ""
tables["Grocery Stores"] = ""
```

```python
tables.head()
```

```python
# create a params dict that will be updated with new city each iteration
params = {"key": "AIzaSyBiTRcGCSJDHIgUPNKGVzTBYmd6uliMzRU"}

# Loop through the cities_pd and run a lat/long search for each state
for index, row in tables.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    state = row['State']

    # update address key value
    params['address'] = f"{state}"

    # make request
    state_lat_lng = requests.get(base_url, params=params)
    
    # convert to json
    state_lat_lng = state_lat_lng.json()

    tables.loc[index, "Latitude"] = state_lat_lng["results"][0]["geometry"]["location"]["lat"]
    tables.loc[index, "Longitude"] = state_lat_lng["results"][0]["geometry"]["location"]["lng"]

# Visualize to confirm lat lng appear
tables.head()
```

**Extract data on California stores**

```python
# referenced from https://stackoverflow.com/questions/46578128/pandas-read-jsonjson-url
from urllib.request import urlopen
```

```python
CA_stores = urlopen("https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=36.7783,%20-119.418&radius=50000&keyword=Grocery%20store&key=AIzaSyBiTRcGCSJDHIgUPNKGVzTBYmd6uliMzRU")
json_data = CA_stores.read()
```

```python
d = json.loads(json_data)
```

```python
df = pd.json_normalize(d['results'])
```

```python
df.head() 
```

```python
df = df.drop(columns = ['business_status', 'icon', 'photos', 'place_id', 'rating', 'reference', 'scope', 'types', 'user_ratings_total', 'geometry.viewport.northeast.lat', 'geometry.viewport.northeast.lng', 'geometry.viewport.southwest.lat', 'geometry.viewport.southwest.lng', 'opening_hours.open_now', 'plus_code.global_code', 'price_level'])
```

```python
df.head()
```

```python
df[['plus_code.compound_code','State']] = df['plus_code.compound_code'].str.split(',',expand=True)
```

```python
df.head()
```

```python
df = df.drop(columns=['vicinity', 'plus_code.compound_code'])
```

```python
df.head()
```

```python
df.to_csv("california_stores.csv", header=True)
```
